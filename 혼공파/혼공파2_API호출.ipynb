## 파이썬으로 API 호출하기 : requests 패키지
- 웹 브라우저를 사용하지 않고 '도서관 정보나루' 공개 API에서 JSON데이터를 자동으로 추출하는 프로그램을 만들기
  - 매번 브라우저에서 주소+인증키로 JSON 데이터를 다운로드하는 것은 너무 수동적


```python
# API호출 requests패키지 불러오기
import requests
```

```python
# 도서관 정보나루 API인증키 넣은 호출 URL을 변수에 저장
# http://data4library.kr/api/loanItemSrch?format=json&startDt=2023-06-01&endDt=2023-06-30&age=20&authKey=(발급받은 인증키)
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2023-06-01&endDt=2023-06-30&age=20&authKey=6eb67d95a8d0a43198a6c5c1e171b90a47648f7184f3ee96cd7ad45bff491240"
```

```python
# HTTP GET방식
# 웹 브라우저가 웹 서버에 요청 할 때 URL 파라미터 값이나 데이터를 전달하는 방식
# get()함수가 반환하는 값은 API 호출의 결과를 담고 있는 requests패키지의 Response클래스 객체
r = requests.get(url)
```

```python
# json()메서드 : 웹 서버로부터 받은 JSON문자열을 파이썬 객체로 변환하여 반환
data = r.json()
print(data)
```

```python
# print()함수 없어도 코드 셀의 마지막 라인의 실행 결과를 자동으로 출력해주는 코랩, 주피터노트북의 편리한 기능
data
```

### JSON 데이터를 PANDAS 데이터프레임으로 변환 : data딕셔너리 구조 바꾸기

```python
# data['response']['docs']에 매핑된 리스트를 순회하면서 → doc키에 매핑된 딕셔너리 추출한 후 → 빈 리스트에 추가 ▶ for문 이용
books = []
for d in data['response']['docs']:
  books.append(d['doc'])
```

```python
# 위의 방법과 동일
# 파이썬의 리스트 내포(list comprehension)로 만들기
# 리스트 내포(list comprehension) : for문을 대괄호로 감싸서 리스트를 만드는 방법
books = [d['doc'] for d in data['response']['docs']]
```

```python
books
```

### PANDAS DataFrame클래스에 위 리스트를 넘김

```python
import pandas as pd
books_df = pd.DataFrame(books)
books_df
```

```python
# JSON으로 변환하여 저장
books_df.to_json('23s_best_book.json')
```

```python
# JSON으로 변환하여 저장
# to_json()메서드가 한글을 그대로 저장하게 하려면, force_ascii=False지정하기
books_df.to_json('23s_best_book.json', force_ascii=False)
```
