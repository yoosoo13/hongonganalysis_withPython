# 이미지 읽기_imread()함수

```python
import matplotlib.pyplot as plt
img = plt.imread('jupiter.png')
img.shape
```

# 이미지 화면에 출력하기_imshow()함수

```python
plt.imshow(img)
plt.show()
```

```python
# imshow()함수는 피겨 크기 상관없이 기보적으로 원본이미지 가로세로 비율 유지
# 축, 눈금 없애기 : axis()함수를 'off'로 지정
plt.figure(figsize=(8,6))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# 한번에 쿼리 작성

import sys
import os  # os 모듈을 추가합니다.

if 'google.colab' in sys.modules:
    # 샘플 이미지 다운로드
    !wget https://bit.ly/3wrj4xf -O jupiter.png

# 현재 작업 디렉토리를 확인합니다.
print("Current working directory:", os.getcwd())

import matplotlib.pyplot as plt
img = plt.imread('jupiter.png')

# 이미지 플로팅
plt.imshow(img)
plt.axis('off')  # 축 표시 끄기
plt.show()
```

# 이미지 저장하기_imsave()함수
- 첫 번째 매개변수 : 저장할 파일 이름
- 두 번째 매개변수 : 이미지가 저장된 넘파이 배열(파일 이름의 확장자를 사용해 자동으로 이미지를 변환해 줌)

```python
plt.imsave('jupiter.png', img) # 목성 이미지를 JPEG 포맷으로 저장할 수 있음
```
