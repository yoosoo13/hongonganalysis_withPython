# 중간에 있는 '열'삭제하기

```python
# 방법1. 불리언배열
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# 방법2. drop()메서드
ns_book = ns_df.drop('Unnamed: 13', axis=1)
# axis=0: 행을 제거하는 경우
# axis=1: 열을 제거하는 경우
ns_book.head()
```

```python
# 'Unnamed: 13'와 '부가기호' 두 개의 열을 동시에 제외 → 여러 개 열을 제외하려면, 열 이름을 []리스트형식으로 지정
ns_book = ns_df.drop(['Unnamed: 13', '부가기호'], axis=1)
ns_book.head()
```

```python
# inplace매개변수로 현재 선택한 데이터프레임을 바로 수정할 수 있음 → ns_book데이터프레임에 덮어씀
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
# inplace=True일 때 ns_book변수에 연결된 객체가 수정되는 것은 아님. 판다스가 내부적으로 수정된 새로운 객체를 만든 후 ns_book변수에 연결함. 따라서 성능상의 이득은 없음
```

```python
# 방법3. dropna()메서드
# dropna()메서드는 기본적을 NaN(비어있는 값)이 하나 이상 포함된 행이나 열 삭제

# NaN이 포함된 열 삭제해보자
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
# 모든 값이 NaN인 열을 삭제하려면, dropna()메서드에 how매개변수를 'all'로 지정
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

```python
# inplace=True를 지정하여 ns_book데이터프레임을 수정가능
ns_book.dropna(axis=1, how='all', inplace=True)
ns_book.head()
```

# 중복된 행 찾기
- duplicated()메서드 사용

```python
# 모든 '열'기준으로 중복된 행 찾기
sum(ns_book.duplicated())
# 현재는 ns_book 데이터프레임의 모든 행은 '번호'열에 고유한 값을 가지므로 중복된 행이 없음
```

```python
# 일부 열을 기준으로 중복된 행 찾으려면, duplicated()메서드의 subset매개변수 사용
# '도서명', '저자', 'ISBN'을 기준으로 중복된 행 찾기
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
```

```python
# duplicated()메서드에 keep매개변수=False지정하여 중복된 모든 행을 True로 표시
# '도서명', '저자', 'ISBN'을 기준으로 어떤 데이터가 중복되었는지 확인
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

# 그룹별로 모으기
- 앞으로 어떤 도서가 인기 있을지 예상하려고 함 > ns_book데이터프레임에서 '대출건수'열이 중요
- groupby()메서드 사용

```python
# 데이터프레임 합치기
# 그룹으로 묶을 기준 열 : ['도서명', '저자', 'ISBN', '권']
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
```

1. groupby()메서드는 기본적으로 by 매개변수에 지정된 열에 NaN이 포함되어 있으면 해당 행을 삭제함
2. count_df 데이터프레임의 '도서명', '저자', '권'열에는 값이 누락되어 이따금 NaN이 포함되어 있음
3. NaN이 포함되어 있는 행을 삭제하고 계산하면, 대출건수 합계에서 빠짐...
4. 3번의 대출건수 합계에서 빠지는걸 막기 위해, 'dropna 매개변수 = False'로 지정
  - dropna = False는 NaN이 있는 행도 포함하겠다는 의미

```python
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False)
loan_count = group_df.sum()
loan_count.head()
```

```python
# 두 메서드 이어서 작성
loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna = False).sum()
loan_count.head()

# 인덱스 열은 굵은 글씨로 표시
```

# 원본 데이터 업데이트하기
- 더해진 '대출건수'를 업데이트하기 전 전처리 과정(원본 데이터프레임에는 중복된 데이터가 존재)
  1. duplicated() 메서드로 중복된 행을 True로 표시한 불리언 배열 만들기
  2. 1번에서 구한 불리언 배열을 반전시켜서 중복되지 않은 고유한 행을 True로 표시
  3. 2번에서 구한 불리언 배열을 사용해 원본 배열에서 고유한 행만 선택

```python
# 1. 중복된 행을 True로 표시
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])

# 2. 불리언 배열을 반전시켜 고유한 행을 True로 표시
# ~연산자로 불리언 배열을 반전
unique_rows = ~dup_rows

# 3. 고유한 행만 선택
# copy()메서드 :  원본 데이터의 복사본을 생성(독립적으로 메모리 할당하여, 원본데이터에는 영향미치지 않음)
ns_book3 = ns_book[unique_rows].copy()
```

```python
# 중복된 행 존재여부 확인
sum(ns_book3.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

## 원본 데이터프레임 인덱스 설정하기
- ns_book3의 인덱스를 loan_count 데이터프레임의 인덱스와 동일하게 만듭니다.
  - 지정한 열을 인덱스로 설정할 때 : set_index()메서드 사용
  - Inplace = True로 지정해, 새로운 데이터 프레임 반환하지 않고 ns_book3데이터프레임 수정

```python
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.head()
```

## 업데이트하기: update()메서드
- 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트할 때 쓰는 메서드
- reset_index()메서드 : 업데이트가 제대로 되었다면, 인덱스 열을 해제

```python
ns_book3.update(loan_count)
ns_book3.head()
```

```python
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
# ns_book에서 '대출건수'가 100회 이상인 책 갯수 세어보기
sum(ns_book['대출건수'] >= 100)
```

```python
# 열 순서 정렬 : []연산자 사용
ns_book4 = ns_book4[ns_book.columns] # ns_book데이터프레임의 열 이름 전달
ns_book4.head()
```

# 일괄 처리 함수 만들기(요약)

```python
def data_cleaning(filename):
# 남산 도서관 장서 CSV 데이터 전처리 함수
# :param filename" CSV 파일 이름

  # 파일을 데이터프레임으로 읽습니다.
  ns_df = pd.read_csv(filename, low_memory = False)

  # NaN인 열을 삭제합니다.
  ns_book = ns_df.dropna(axis=1, how='all')

  # 대출건수를 합치기 위해 필요한 행만 추출하여 count_df 데이터프레임을 만듭니다.
  count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]

  # 도서명, 저자, ISBN, 권을 기준으로 대출건수 groupby합니다.
  loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()

  # 원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사합니다.
  duo_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()

  # 도서명, 저자, ISBN,권을 인덱스로 설정합니다.
  ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)

  # load_count에 저장된 누적 대출건수를 업데이트합니다.
  ns_book3.update(loan_count)

  # 인덱스를 재설정합니다.
  ns_book4 = ns_book3.reset_index()

  # 원본 데이터프레임의 열 순서로 변경합니다.
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4
```

```python
# 데이터프레임 비교 : equals()메서드 사용
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4)
```
