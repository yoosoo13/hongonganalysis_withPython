## 누락된 값 바꾸기(1) : loc, fillna()메서드
- '세트 ISBN'열은 대부분 비어 있다.
  - loc메서드 : 누락된 값을 NaN이 아니라, 빈 문자열('')로 바꾸기

```python
# 1. 누락된 값을 찾아 불리언 배열로 반환_열의 값들이 True, False로 변환됨
set_isbn_na_rows = ns_book4['세트 ISBN'].isna()

# 2. 누락된 값을 빈 문자열('')로 바꾸기
ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''

# 3. 누락된 값이 몇 개인지 세기
ns_book4['세트 ISBN'].isna().sum()
```

```python
# fillna()메서드 : 원하는 값을 전달하면 NaN을 대체할 수 있음
# ns_book4에 있는 모든 NaN을 '없음'문자열로 바꾸기
ns_book4.fillna('없음').isna().sum()
```

```python
# 특정 열만 선택해서 NaN로 바꿀 수 있음
ns_book4['부가기호'].fillna('없음').isna().sum()

# 특정 열만 선택해서 fillna()메서드를 적용하면, 열 이름 없이 개수만 있는 판다스 시리즈 객체로 반환(=열 이름이 없이 각 행의 값들만이 출력되는 것)
```

## 누락된 값 바꾸기(2): replace()메서드
- replace()메서드는 NaN는 물론 어떤 값도 바꿀 수 있는 편리한 메서드

```python
# 1. 바꾸려는 값이 하나일 때, replace(원래 값, 새로운 값)
ns_book4.replace(np.nan, '없음').isna().sum()
```

```python
# 2-1. 바꾸려는 값이 여러개 일 때, replace([원래 값1, 원래 값 2], [새로운 값1, 새로운 값2])
ns_book4.replace([np.nan, '2021'], ['없음', '21']).head(2)
```

```python
# 2-2. 바꾸려는 값이 여러개 일 때, replace({원래 값1 : 새로운 값1, 원래 값2 : 새로운 값2})
ns_book4.replace({np.nan:'없음', '2021':'21'}).head(2)
```

```python
# 3. 열 마다 다른 값으로 바꿀 때, replace({열 이름 : 원래 값}, 새로운 값)
# 예. '부가기호'열의 NaN을 '없음'으로 바꾸기
ns_book4.replace({'부가기호':np.nan}, '없음').head(2)
```

```python
ns_book4.replace({'부가기호' : {np.nan:'없음'},
                  '발행년도' : {'2021':'21'}}).head(2)
```

## 숫자 찾기 : \d
- 네 자리 연도 : \d\d\d\d
- 네 자리 연도에서 뒤에 두 자리만 하나의 그룹으로 묶을 때 : \d\d(\d\d)
- \d\d == \d{2}
- \d\d(\d\d) == \d{2}(\d{2})

```python
# '발행년도' 열의 값을 정규 표현식으로 두 자리 연도로 바꾸기 -- regex매개변수 사용
ns_book4.replace({'발행년도' : {r'\d\d(\d\d)' : r'\1'}}, regex=True)[100:102]
```

```python
ns_book4.replace({'발행년도' : {r'\d{2}(\d{2})' : r'\1'}}, regex=True)[100:105]
```

```python
def get_book_title(isbn):
  # Yes24 도서 검색 페이지 URL
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
  # URL에 ISBN을 넣어 HTML 가져오기
  r= requests.get(url.format(isbn))
  # HTML 파싱
  soup = BeautifulSoup(r.text, 'html.parser')
  # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옵니다.
  title = soup.find('a', attrs={'class' : 'gd_name'}).get_text()
  return title
```
