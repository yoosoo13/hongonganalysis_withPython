## (그래프 그리는 방법)-pyplot방식과 객체지향 API방식
- matplotlib.pylot에 있는 pyplot방식
- 객체지향 API : 피겨 객체와 서브플롯 객체를 만들고 이 객체의 메서드를 사용


````python
# 그래프가 잘 보이도록 해상도 높여주기
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 100
````

### pyplot방식으로 그래프 그리기
- matplotlib.pyplot에 있는ㄴ 함수를 사용하면, 함수들이 하나의 피겨 객체에 대한 상태를 공유
- 즉, 간단한 선 그래프를 그리면 plot()함수와 title()함수는 동일한 피겨 객체에 적용됨

````python
# plot()함수에 리스트 형태로 데이터를 전달 → title()함수에는 그래프 제목을 넣어 그려보기
plt.plot([1,4,9,16])
plt.title('simple line graph')
plt.show()
````

```
 - 만약 하나의 리스트만 전달한마녀 plot()함수는 원소를 y축의 값이라고 인식
 - 이 리스트의 인덱스를 x축의 값으로 사용합니다.
 - 따라서 위 예시의 경우 [1,4,9,16]은 y축 값이며 x축의 값은 [0,1,2,3]이 됩니다.
 ```

### 객체지향 API방식으로 그래프 그리기


````python
fig, ax = plt.subplots()   # 하나의 Axes 객체를 가지는 피겨를 생성합니다.
ax.plot([1,4,9,16])
ax.set_title('simple line graph')
fig.show()
````

## 출판사별 발생 도서 개수 산점도 그리기
- ns_book7 데이터프레임을 사용해서 출판사별 도서가 어떤 연도에 많이 발행되었는지 산점도 작성

### 데이터 불러오기

````python
import gdown
gdown.download('http://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
````

```
# 연도를 기준으로 출판사별 발행 도서 개수를 표현
# x축-발행년도, y축-출판사 표현
# 전체 출판사 수가 2만개가 넘으므로, 발행도서가 많은 상위 30개 출판사 데이터 중 일부를 사용
```

### 상위 30개 출판사 선택

````python
# 상위 30개 출판사 선택 : 고유한 출판사 목록 만들기-value_counts()메서드 사용(내림차순으로 정렬됨)
top30_pubs = ns_book7['출판사'].value_counts()[:30]
top30_pubs
````

````python
# 상위 30개 출판사에 해당하는 행을 표시하는 불리언 인덱스 만들기 : isin()메서드 사용
top30_pubs_idx = ns_book7['출판사'].isin(top30_pubs.index)
top30_pubs_idx
````

```
isin(top30_pubs.index) : 상위 30개에 해당하는 출판사는 True, 아닌 출판사는 False로 반환
```

````python
# 상위 30개 출판사의 발행 도서 개수가 총 몇개?
top30_pubs_idx.sum()
````

### 데이터프레임 랜덤으로 1000개로 설정
- random_state는 호출할 때마다 동일한 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값
- 랜덤(sample(데이터개수 작성, random_state=42))지정

````python
# 5만개가 넘는 데이터, 산점도로 보기엔 여전히 데이터가 너무 많다. 이 중에서 1,000개만 선택해보기: sample()메서드 사용-데이터프레임의 행을 무작위로 선택
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
````

````python
# 산점도 그리기(x축-'발행년도', y축-'출판사')
# 조금 크게 그리기 위해 subplots()함수로 피겨의 크기를 (10,8)로 지정 - 객체지향API방식으로 그리기
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'])
ax.set_title('출판사별 발행 도서')
fig.show()
````

## 마커꾸미기
  1. 투명도 조절: alpha 매개변수
  2. 마커 테두리 색 변경: edgecolor 매개변수(기본값-'face')
  3. 마커 테두리 선 두께 변경: linewidths 매개변수(기본값-1.5)
  4. 산점도 색 변경: c 매개변수
    - 데이터 개수와동일한 길이의 배열을 전달하면, 각 데이터를 다른 색깔로 설정 할 수 있음
  5. 마커 크기 설정: s 매개변수

````python
# 값에 따라 마커 크기 다르게 표현: s 매개변수
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], s = ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
````

```
- 각 출판사의 연도별 도서 발행 여부뿐만 아니라 얼마나 많은 대출을 기록했는지 확인가능
- 2000년 초반에는 '자음과모음'출판사에서 발행한 책의 대출건수가 높음을 확인
- 2005~2010년에는 '김영사'출판사에서 발행한 도서의 대출건수가 높음을 확인
```

````python
# 마커 꾸미기
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'],
           linewidths=0.5 ,edgecolor = 'k', alpha = 0.3,
           s = ns_book8['대출건수']*2, c = ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
````

## 값에 따라 색상 표현하기: 컬러맵
- 컬러맵: cmap 매개변수 사용
- 앞서 그린 산점도는 기본값: viridis컬러맵으로 표현됨
- 자주 사용하는 컬러맵: jet컬러맵
  - jet컬러맵 낮은 값일수록, 짙은 파란색
  - jet컬러맵 높은 값일수록, 노란색 → 붉은색

```
컬러맵 레퍼런스
- https://matplotlib.org/stable/gallery/color/colormap_reference.html
```

- 컬러막대: colorbar() 메서드 사용

````python
# 컬러맵: cmap 매개변수 사용, 현재 jet컬러맵으로 작성

# 컬러막대: scatter()함수가 반환하는 객체를 colorbar()메서드에 전달
# 컬러맵의 색깔이 어떤 대출건수 값에 대응하는지 참조정보 제공하는 컬러막대를 그래프 옆에 작성

fig, ax = plt.subplots(figsize=(10,8))
sc = ax.scatter(ns_book8['발행년도'], ns_book8['출판사'],
           linewidths=0.5, edgecolor='k', alpha = 0.3,
           s=ns_book8['대출건수']*1.3, c=ns_book8['대출건수'], cmap='jet')
ax.set_title('출판사별 발행 도서')
fig.colorbar(sc)
fig.show()
````

```
1. 대출건수가 많을수록 마커의 크기가 커짐 > 마커 크기를 조절하면 산점도에 부가적인 정보를 포함시킬 수 있다.
```
