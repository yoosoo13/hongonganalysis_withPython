# 1-2_b.시도별 5대 범죄간의 상관관계
- 각 시도 발생건수를 합쳐 대한민국 5대범죄현황으로 데이터프레임을 생성

- 5대범죄 변수들 간의 관계를 산점도로 시각화
  - 서울, 부산, 대구 등 각 지역의 데이터프레임을 리스트에 저장하고 for 루프를 사용하여 각 지역별 히트맵 생성
  - 각 지역의 이름을 제목으로 설정하여 각 히트맵을 구분

```python
import random

# 서울, 부산, 대구, ... , 제주 데이터프레임을 리스트에 저장
city_dfs = [seoul_df, busan_df, deajeon_df, deagu_df, ulsan_df, incheon_df, gwangju_df, gyeonggido_df, gangwondo_df, chungbuk_df, chunganam_df, jeollabukdo_df, jeollanamdo_df, gyeongbuk_df, gyeongnam_df]
city_names = ['서울', '부산', '대전', '대구', '울산', '인천', '광주', '경기도', '강원도', '충북', '충남', '전라북도', '전라남도', '경북', '경남']

fig, axs = plt.subplots(nrows=5, ncols=3, figsize=(12,18))

# 색상 맵 설정
colors = ['Blues', 'Reds', 'Greens', 'Purples', 'Oranges', 'Greys', 'YlOrBr', 'YlGnBu', 'YlOrRd', 'BuPu', 'RdPu', 'PuRd', 'RdYlGn', 'Pastel1', 'Pastel2', 'Set1', 'Set2']

for i, df in enumerate(city_dfs):
    row = i // 3
    col = i % 3

    # 랜덤한 RGB 색상 생성
    color = colors[i % len(colors)]

    sns.heatmap(data=df[['살인', '강도', '성범죄', '절도', '폭력']].corr(), annot=True, cmap=color, ax=axs[row, col])
    axs[row, col].set_title(city_names[i])

# 마지막 subplot은 빈 subplot으로 남깁니다.
axs[4, 2].axis('off')

plt.tight_layout()
plt.show()
```

# 1-2_c. 2020년 기준, 대한민국 5대 범죄간의 상관관계
- 전국의 5대범죄(살인, 강도, 성범죄, 절도, 폭력)에 대한 상관계수를 히트맵으로 시각화
- 색상은 RdPu로 지정되어 있으며, 이는 높은 상관관계 값을 보여주는 셀이 보라색에 가깝게 표시

```python
# 전체 데이터 프레임을 만들기 위해 모든 지역의 데이터 프레임을 합침
total_df = pd.concat(city_dfs)

# 히트맵 그리기
plt.figure(figsize=(8,6))

# 이전 코드와 마찬가지로 히트맵을 그려주는데, annot=True로 설정하여 각 셀에 해당하는 상관계수 값을 표시
sns.heatmap(data=total_df[['살인', '강도', '성범죄', '절도', '폭력']].corr(), annot=True, cmap='RdPu')

# 제목 작성
plt.title('2020년 대한민국 5대범죄 범죄 상관관계')

plt.show()
```

# 2-1. 서울시 5대 범죄 발생건수와 서울시 유흥주점 개수 간의 상관관계 분석

- 범죄에 관련있는 요인으로 '유흥주점'을 선정한 이유
  - 유흥주점은 사회적인 모임과 야간 활동의 장으로서 범죄 발생과 관련이 있을 수 있기 때문에 범죄에 관련된 요인으로 선정
- 17개 시도 중 '서울' 선정한 이유
 -  유흥주점관련 공공데이터가 다른 시도에는 생략되어있는 정보가 많았던 가운데 '서울'은 '서울시 관할구'에 대한 정보가 모두 존재했기 때문에 서울시를 선정

```python
# 서울시 유흥업소 데이터 전처리 요약
# 2020년도 유흥주점 개수를 count 하기 위해, '영업중+인허가 일자가 2020년 이하인 데이터'를 필터링 ▶ 엑셀의 countif 함수를 이용해 서울시의 각 자치구 별 유흥주점의 개수를 구하여 서울시 5대 범죄 인구 데이터에 열을 추가함

#  서울시 유흥업소 데이터 불러오기
url2="https://drive.google.com/file/d/1ENFWh-LK5_M2tDytCziH96HlYn_G8-af/view?usp=sharing"
path2='https://drive.google.com/uc?id='+url2.split('/')[-2]
print("▶ 공유파일 링크변환 경로명 : ", path2)
s_df=pd.read_csv(path2)
s_df.head()
```
